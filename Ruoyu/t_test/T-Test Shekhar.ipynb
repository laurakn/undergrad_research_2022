```python
from scipy.stats import ttest_ind
import pandas as pd
```

```python
lower_bound = 0.4
upper_bound = 0.6

def process_df(df):
    df = df[(df['checkpoint'] == 1000)]
    df = df[(df['C'] >= lower_bound)]
    df = df[(df['C'] <= upper_bound)]
    return df
```

```python
df_R = pd.read_csv("Random_Sampling_1000_R.csv")
#df_R = process_df(df_R)
```

```python
display(df_R)
```

```python
df_L = pd.read_csv("LogReg_baseline_1000_L.csv")
#df_L = process_df(df_L)
df_L
```

```python
alpha = 0.05/len(df_L['C'].unique())
for c in df_L['C'].unique():
    tmp = df_R[df_R['C'] == c]
    R_cmp = tmp['Equal_Opportunity'].to_numpy()
    
    tmp = df_L[df_L['C'] == c]
    L_cmp = tmp['equal_opportunity'].to_numpy()
    #print(R_cmp)
    #print(L_cmp)
    #print(c)
    stats, p = ttest_ind(R_cmp, L_cmp)
    print(f'For C= {c}, Statistics=%.3f, p=%.3f' % (stats, p))
    if p > alpha:
        print('Same distribution')
    else:
        print('Different distribution')
```
